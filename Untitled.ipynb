# Data Cleaning

## Handling Missing Values 

```python
# Handle missing values
print("### Handling Missing Values ###")
columns=['Name','Anthony','SibSp','Cabin','Embarked']
for column in df.columns:
    if df[column].dtype == 'object':  # Check if the column is categorical
        df[column].fillna(df[column].mode()[0], inplace=True)  # Fill missing values with the mode (most frequent value)
    else:  # If the column is numerical
        df[column].fillna(df[column].median(), inplace=True)  # Fill missing values with the median
print(df)
```

### Verifying Missing values are handled

```python
# Verify missing values are handled
print("\nMissing values after handling:")
print(df.isnull().sum())  # Print the number of missing values per column after handling
```

#### Check for potential errors by examining unique values and data types

```python
# Check for potential errors by examining unique values and data types
print("### Checking for Potential Errors ###")

print("\nData types of columns:")
print(df.dtypes)  # Print data types of each column
for column in df.columns:
    unique_values = df[column].unique()  # Get unique values in the column
    print(f"\nUnique values in column '{column}':")
    print(unique_values)  # Print unique values in the column
```

### Check the range of values for numeric columns

```python

print("### Checking Range of Values for Numeric Columns ###")
for column in df.select_dtypes(include=['int64', 'float64']).columns:
    min_value = df[column].min()  # Get the minimum value in the column
    max_value = df[column].max()  # Get the maximum value in the column
    print(f"\nValue range in column '{column}': {min_value} to {max_value}")  # Print the range of values
```

###### Checking for  Outliers using box plot 

```python
# Check for outliers using box plot
print("### Checking for Outliers using Box Plot ###")
for column in df.select_dtypes(include=['int64', 'float64']).columns:
    plt.figure(figsize=(10, 6))  # Set the figure size for the plot
    sns.boxplot(x=df[column])  # Create a box plot for the column
    plt.title(f'Box plot of {column}')  # Set the title of the box plot
    plt.show()  # Display the box plot
```

#### Checking for outliers using z-Score method

```python
# Check for outliers using Z-score
print("### Checking for Outliers using Z-score ###")
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns  # Get numeric columns
z_scores = df[numeric_cols].apply(zscore)  # Calculate Z-scores for numeric columns
outliers_z = (abs(z_scores) > 3).any(axis=1)  # Identify rows with Z-scores greater than 3
print("\nOutliers detected by Z-score method:")
print(df[outliers_z])  # Print rows detected as outliers by Z-score
```

##### Handling outliers 

```python
# Handle outliers by capping them at 3 standard deviations
print("### Handling Outliers ###")
columns=['Name','Sex','Ticket','Cabin','Embarked']
for column in numeric_cols:
    # Cap outliers at 3 standard deviations from the mean
    df[column] = df[column].apply(lambda x: x if abs(x - df[column].mean()) / df[column].std() <= 3 else df[column].mean())
```

##### Verifying  outliers handling

```python

z_scores = df[numeric_cols].apply(zscore)  # Recalculate Z-scores after handling outliers
outliers_z = (abs(z_scores) > 3).any(axis=1)  # Identify rows with Z-scores greater than 3
print("\nOutliers after handling by Z-score method:")
print(df[outliers_z])  # Print rows detected as outliers by Z-score after handling
```

# Data Transformation

## Convert categorical data into numeric format using one-hot encoding

```python

print("### Converting Categorical Data into Numeric Format ###")
df = pd.get_dummies(df, columns=['Name','Sex','Ticket','Cabin','Embarked'],dtype="float")  # Apply one-hot encoding to categorical columns

print("\nData after one-hot encoding:")
print(df.head())  # Display the first few rows of the dataframe after one-hot encoding


```

## Normalize or standardize the numerical values if required for later modeling

```python
# Normalize or standardize the numerical values if required for later modeling

print("### Normalizing or Standardizing Numerical Values ###")
numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns  # Get numeric columns
scaler = MinMaxScaler()  # Initialize the MinMaxScaler
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])  # Apply normalization to numeric columns

print("\nData after normalization:")
print(df.head())  # Display the first few rows of the dataframe after normalization
```
